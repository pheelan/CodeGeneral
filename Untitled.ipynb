```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
I = 20  # Current in Amps
rho_copper = 1.68e-8  # Resistivity of copper (Ohm meter)
A = 3e-6  # Cross-sectional area of wire (m^2)
L = 1  # Length of wire (1m assumed for simplicity)
T_ambient = 25  # Ambient temperature in Celsius
k_insulation = 0.2  # Thermal conductivity of insulation (W/mK) typical for rubber or PVC
h_air = 10  # Convection coefficient (W/m^2K), rough estimate

# Resistance of wire
R = rho_copper * L / A
P = I**2 * R  # Joule heating power in Watts

# Insulation thickness range (0 to 10mm)
thickness = np.linspace(0, 0.01, 100)  # 0mm to 10mm (converted to meters)

# Compute wire temperature for each insulation thickness
temperatures = []
for t in thickness:
    if t == 0:
        # No insulation, only convection
        thermal_resistance = 1 / (h_air * np.pi * (0.002) * L)  # Bare wire convective resistance
    else:
        r_inner = 0.001  # Wire radius (2mm diameter)
        r_outer = r_inner + t  # Outer radius with insulation
        
        # Thermal resistance of insulation (cylindrical conduction)
        R_insulation = np.log(r_outer / r_inner) / (2 * np.pi * k_insulation * L)
        
        # Outer convection resistance
        R_convection = 1 / (h_air * (2 * np.pi * r_outer * L))
        
        # Total thermal resistance
        thermal_resistance = R_insulation + R_convection
    
    # Steady-state temperature
    T_wire = T_ambient + P * thermal_resistance
    temperatures.append(T_wire)

# Plot results
plt.figure(figsize=(8,5))
plt.plot(thickness * 1000, temperatures, label='Wire Temperature', color='r')
plt.xlabel('Insulation Thickness (mm)')
plt.ylabel('Wire Temperature (°C)')
plt.title('Effect of Insulation Thickness on Wire Temperature')
plt.grid()
plt.legend()
plt.show()
```

```python


# Constants
I = 20  # Current in Amps
rho_copper = 1.68e-8  # Resistivity of copper (Ohm meter)
A = 3e-6  # Cross-sectional area of wire (m^2)
L = 1  # Length of wire (1m assumed for simplicity)
T_ambient = 25  # Ambient temperature in Celsius
k_insulation = 0.1  # Thermal conductivity of insulation (W/mK)
rho_insulation = 1200  # Density of insulation (kg/m^3)
c_insulation = 1500  # Specific heat capacity of insulation (J/kgK)
h_air = 5  # Natural convection coefficient (W/m^2K)

# Resistance of wire
R = rho_copper * L / A
P = I**2 * R  # Joule heating power in Watts

# Time parameters
time_steps = 100000  # Number of time steps
dt = 1  # Time step size (seconds)

# Spatial parameters (for insulation layers)
thickness = 0.01  # 10mm insulation
N_layers = 10  # Number of discrete insulation layers
dr = thickness / N_layers  # Radial step size
radii = np.linspace(0.001, 0.001 + thickness, N_layers)  # Radii from wire surface to outer surface

# Initialize temperature profile
T = np.full(N_layers, T_ambient)
T_wire = T_ambient

# Thermal diffusivity of insulation
alpha = k_insulation / (rho_insulation * c_insulation)

# Store temperature evolution
T_wire_history = []
T_outer_history = []
time = np.arange(0, time_steps * dt, dt)

# Time-stepping loop
for _ in range(time_steps):
    T_new = T.copy()
    
    # Update temperature of insulation layers using explicit finite difference
    for i in range(1, N_layers - 1):
        d2T_dr2 = (T[i+1] - 2*T[i] + T[i-1]) / dr**2
        T_new[i] = T[i] + alpha * d2T_dr2 * dt
    
    # Boundary condition at the inner surface (wire heating)
    heat_flux = P / (4 * np.pi * radii[0]**2 * L)  # Heat flux per unit area
    T_new[0] = T_wire + heat_flux * dr / k_insulation
    
    # Boundary condition at the outer surface (convective cooling)
    T_new[-1] = T[-1] - h_air * (T[-1] - T_ambient) * dt / (rho_insulation * c_insulation * dr)
    
    # Update temperature profile
    T = T_new
    T_wire = T[0]
    
    # Store history
    T_wire_history.append(T_wire)
    T_outer_history.append(T[-1])

# Plot transient temperature evolution
plt.figure(figsize=(8,5))
plt.plot(time, T_wire_history, label='Wire Temperature', color='r')
plt.plot(time, T_outer_history, label='Outer Insulation Temperature', color='b')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Transient Heating of Wire and Insulation')
plt.grid()
plt.legend()
plt.show()
```
