```python
    # Combine the path with the file name
    file_name = 'BatteryData-3376.txt'
    full_file_path = test_file_path / file_name

    # Load the file with pandas
    data = pd.read_csv(full_file_path, delim_whitespace=True, header=None)

    # Print the first few rows
    print(data.head())
```

```python
# Convert Unix time (column 0) to datetime

data['datetime'] = pd.to_datetime(data[0], unit='s')

# Split datetime into separate columns for date and time
data['date'] = data['datetime'].dt.date  # Extract the date part
data['time'] = data['datetime'].dt.time  # Extract the time part

# Drop the original datetime column if not needed
data = data.drop(columns=['datetime'])

# Subtract the minimum Unix time to get seconds
data['seconds'] = data[0] - data[0].min()
```
